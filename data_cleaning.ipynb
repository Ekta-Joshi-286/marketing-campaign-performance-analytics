```python
import pandas as pd
```

```python
# Load Data
df = pd.read_csv("marketing_analytics_portfolio_dataset_50k.csv")
```

```python
# Basic Info
print(df.shape)
print(df.info())
```

```python
# Missing Values
print(df.isnull().sum())
```

```python
# Duplicate Records
print("Duplicates:", df.duplicated().sum())
```

```python
# Remove Duplicates
df = df.drop_duplicates()
```

```python
# Convert Date
df['date'] = pd.to_datetime(df['date'])
```

```python
# Check Numeric Columns
print(df.describe())
```

```python
# Save Clean Data
df.to_csv("marketing_cleaned.csv", index=False)
```
