```python
import pandas as pd
```

```python
df = pd.read_csv("marketing_cleaned.csv")
```

```python
# CTR
df['CTR'] = (
    df['clicks']
    /
    df['impressions']
) * 100
```

```python
# Lead Rate
df['Lead_Rate'] = (
    df['leads']
    /
    df['clicks']
) * 100
```

```python
# Conversion Rate
df['Conversion_Rate'] = (
    df['conversions']
    /
    df['leads']
) * 100
```

```python
# Profit
df['Profit'] = (
    df['revenue']
    -
    df['spend']
)
```

```python
# ROI
df['ROI'] = (
    (
        df['revenue']
        -
        df['spend']
    )
    /
    df['spend']
) * 100
```

```python
# Top Campaigns
top_campaigns = (
    df.groupby('campaign_id')['revenue']
    .sum()
    .sort_values(ascending=False)
)

print(top_campaigns.head())
```

```python
# Save Final Dataset
df.to_csv(
    "marketing_final.csv",
    index=False
)

print("Data Transformation Completed")
```
