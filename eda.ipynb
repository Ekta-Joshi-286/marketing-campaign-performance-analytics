```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("marketing_cleaned.csv")
```

```python
# Dataset Overview
print(df.head())
print(df.shape)
```

```python
# Revenue Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['revenue'], bins=30)
plt.title("Revenue Distribution")
plt.show()
```

```python
# Spend Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['spend'], bins=30)
plt.title("Spend Distribution")
plt.show()
```

```python
# Revenue by Channel
channel_rev = (
    df.groupby('channel')['revenue']
    .sum()
    .sort_values(ascending=False)
)

channel_rev.plot(kind='bar')
plt.title("Revenue by Channel")
plt.show()
```

```python
# Revenue by Region
region_rev = (
    df.groupby('region')['revenue']
    .sum()
    .sort_values(ascending=False)
)

region_rev.plot(kind='bar')
plt.title("Revenue by Region")
plt.show()
```

```python
# Revenue by Device
device_rev = (
    df.groupby('device')['revenue']
    .sum()
)
device_rev.plot(kind='pie', autopct='%1.1f%%')
plt.title("Revenue by Device")
plt.show()
```

```python
# Correlation
plt.figure(figsize=(10,6))
sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='Blues'
)
plt.title("Correlation Matrix")
plt.show()
```
